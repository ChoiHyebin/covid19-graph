```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel("C:/Users/computer12/Documents/22covid.xlsx")
data.head()

x = data.loc[:, "date"]
y = data.loc[:, "new_cases"]

fig, ax = plt.subplots(figsize=(30,16))

plt.bar(x, y, label="Cases", color="orange", alpha=0.4)

data["date"] = data["date"]
data["MA5"] = data["new_cases"].rolling(window=5).mean()
data["MA60"] = data["new_cases"].rolling(window=85).mean()

def find_intersections_rising(x, y0, y1, xrange=(None, None)):
    """
    Finds the intersections of y0(x) and y1(x) such that:
        1) y0(x) = y1(x) = y
        2) d/dx (y0(x) - y1(x)) > 0
        3) x is in xrange
    xrange defaults to the entire range of the data by default.
    Returns (x, y).
    """
    xrange = (
        x.min() if xrange[0] is None else xrange[0],
        x.max() if xrange[1] is None else xrange[1],
    )

    x = np.asanyarray(x)
    y0 = np.asanyarray(y0)
    y1 = np.asanyarray(y1)

    idxs = np.argwhere((xrange[0] <= x) & (x <= xrange[1])).squeeze()
    x = x[idxs]
    y0 = y0[idxs]
    y1 = y1[idxs]

    # https://stackoverflow.com/a/46911822
    def find_roots(x, y):
        s = np.abs(np.maximum(np.diff(np.sign(y)), 0)).astype(bool)
        return x[:-1][s] + np.diff(x)[s] / (np.abs(y[1:][s] / y[:-1][s]) + 1)

    z = find_roots(x, y0 - y1)

    return z, np.interp(
        np.asanyarray(z, dtype=np.float64),
        np.asanyarray(x, dtype=np.float64),
        np.asanyarray(y0, dtype=np.float64),
    )

plt.plot(data["date"], data["MA5"], label="MA5", alpha=0.8, color="green", linewidth=5)
plt.plot(data["date"], data["MA60"], label="MA60", alpha=0.8, color="blue", linewidth=5)

zxs, zys = find_intersections_rising(
    data["date"],
    data["MA5"],
    data["MA60"],
    xrange=(np.datetime64("2022-04"), np.datetime64("2022-08"))
)

for zx, zy in zip(zxs, zys):
    xd=zx.astype('datetime64[D]')
    plt.plot(zx, zy+20000, color="red", marker="v", markersize=50)
    plt.text(zx, zy+45000, xd, fontsize=30)

plt.xlabel("Date", size=50, rotation=0, labelpad=30)
plt.ylabel("New_Cases", size=50, rotation=0, labelpad=80)

ax.tick_params(axis='x', direction='in', labelsize=30, width=3, pad=3, length=6)
ax.tick_params(axis='y', direction='in', labelsize=30, width=3, pad=3, length=6)

ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

plt.title("covid 19 cases", size=50)
plt.legend(fontsize=30)
plt.grid(True)
plt.show()
```
